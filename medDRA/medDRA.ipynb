```python
df['MedDRA Code'] = df['MedDRA Code'].astype('category')
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop('Unnamed: 3', axis = 1)
```

**LLT to PT**

```python
df_tmp = df[df['Level'] == 'LLT']
df_tmp = df_tmp[['MedDRA Term', 'PT']]
df_tmp['PT'] = df_tmp['PT'].str.upper()
df_tmp['MedDRA Term'] = df_tmp['MedDRA Term'].str.upper()
df_tmp = df_tmp.set_index('MedDRA Term')
dict_1 = df_tmp.to_dict()['PT'] # Converts LLT to PT

#Save it
with open('dict_LLT_PT.pkl', 'wb') as f:
        pickle.dump(dict_1, f, pickle.HIGHEST_PROTOCOL)

dict_1
```
